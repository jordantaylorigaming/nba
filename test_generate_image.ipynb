# Test generate_image Function

This notebook tests the `generate_image` function from `upload_to_server.py`.

The function generates images using Google's Gemini API based on text prompts.


```python
# Import necessary libraries
import os
from dotenv import load_dotenv
from upload_to_server import generate_image
from PIL import Image
import matplotlib.pyplot as plt

# Load environment variables
load_dotenv()

# Get Google API key
google_api_key = os.getenv("GOOGLE_API_KEY")

if not google_api_key:
    print("⚠️ Warning: GOOGLE_API_KEY not found in environment variables")
    print("Please set GOOGLE_API_KEY in your .env file")
else:
    print("✅ Google API key loaded successfully")
```

## Test 1: Generate a simple NBA-related image


```python
# Test with a simple NBA prompt
prompt = "A dynamic basketball court with professional lighting, showing the key and free throw line"
slug = "test-nba-court"

if google_api_key:
    image_path = generate_image(
        prompt=prompt,
        google_api_key=google_api_key,
        slug=slug,
        output_dir="test_images"
    )
    
    print(f"\n{'='*60}")
    print(f"Generated image: {image_path}")
    print(f"{'='*60}")
    
    if image_path and os.path.exists(image_path):
        # Display the image
        img = Image.open(image_path)
        plt.figure(figsize=(12, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Generated Image: {slug}", fontsize=14, pad=20)
        plt.tight_layout()
        plt.show()
        
        print(f"\n✅ Image generated successfully!")
        print(f"📁 Location: {image_path}")
    else:
        print("\n❌ Failed to generate image")
else:
    print("❌ Cannot test - API key not available")
```

## Test 2: Generate multiple images with different prompts


```python
# Test with multiple different prompts
test_cases = [
    ("A minimalist basketball on hardwood court with dramatic shadows", "test-basketball-minimal"),
    ("Basketball arena with audience and court view, professional sports atmosphere", "test-arena-view"),
    ("Close-up of a basketball net with rim, court background blurred", "test-basketball-net"),
]

if google_api_key:
    for i, (prompt, slug) in enumerate(test_cases, 1):
        print(f"\n{'='*60}")
        print(f"Test Case {i}/{len(test_cases)}")
        print(f"Prompt: {prompt}")
        print(f"Slug: {slug}")
        print(f"{'='*60}")
        
        image_path = generate_image(
            prompt=prompt,
            google_api_key=google_api_key,
            slug=slug,
            output_dir="test_images"
        )
        
        if image_path and os.path.exists(image_path):
            print(f"✅ Generated: {image_path}")
            
            # Display the image in a subplot
            img = Image.open(image_path)
            
            # Create subplot for this image
            plt.figure(figsize=(15, 5))
            plt.imshow(img)
            plt.axis('off')
            plt.title(f"Test {i}: {slug}", fontsize=12)
            plt.tight_layout()
            plt.show()
        else:
            print(f"❌ Failed to generate image")
            
        print()  # Empty line for readability
else:
    print("❌ Cannot test - API key not available")
```

## Test 3: Display all generated test images in a grid


```python
# Display all generated images in a grid
import glob

if google_api_key:
    test_images_dir = "test_images"
    
    # Get all PNG files from test_images directory
    image_files = glob.glob(os.path.join(test_images_dir, "*.png"))
    
    if image_files:
        num_images = len(image_files)
        cols = min(3, num_images)
        rows = (num_images + cols - 1) // cols
        
        fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))
        
        # Flatten axes array for easier indexing
        if rows == 1:
            axes = [axes] if cols == 1 else axes
        else:
            axes = axes.flatten()
        
        for idx, image_path in enumerate(image_files):
            if idx < len(axes):
                img = Image.open(image_path)
                axes[idx].imshow(img)
                axes[idx].axis('off')
                axes[idx].set_title(os.path.basename(image_path), fontsize=10)
        
        # Hide extra subplots
        for idx in range(len(image_files), len(axes)):
            axes[idx].axis('off')
        
        plt.tight_layout()
        plt.show()
        
        print(f"✅ Displayed {len(image_files)} test images")
    else:
        print("No test images found")
else:
    print("❌ Cannot display - API key not available")
```

## Test 4: Check function behavior with existing images


```python
# Test that the function returns existing images without regenerating
if google_api_key:
    # Generate an image if it doesn't exist
    test_slug = "test-cache-test"
    prompt = "A basketball resting on a wooden court"
    
    # First call - should generate the image
    print("First call (should generate new image):")
    image_path_1 = generate_image(
        prompt=prompt,
        google_api_key=google_api_key,
        slug=test_slug,
        output_dir="test_images"
    )
    
    print(f"\nSecond call (should use cached image):")
    # Second call - should use existing image
    image_path_2 = generate_image(
        prompt="This prompt is different but slug is same",
        google_api_key=google_api_key,
        slug=test_slug,  # Same slug
        output_dir="test_images"
    )
    
    print(f"\n{'='*60}")
    if image_path_1 == image_path_2 and os.path.exists(image_path_1):
        print("✅ Caching works correctly!")
        print(f"Both calls returned: {image_path_1}")
    else:
        print("❌ Caching may not be working as expected")
    print(f"{'='*60}")
else:
    print("❌ Cannot test - API key not available")
```
